# c1 + c2

```python
import numpy as np
import cv2
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import skimage
import scipy
from scipy.ndimage import gaussian_filter
%matplotlib inline
```

```python
# c1 + c2
img1 = np.float64(cv2.imread('c1.jpg', 0))
a1 = gaussian_filter(img1, sigma=2)

img2 = np.float64(cv2.imread('c2.jpg', 0))
a2 = gaussian_filter(img2, sigma=3)

h = (img1 - a1 + a2) / 2

h = (h - np.min(h)) / (np.max(h) - np.min(h)) * 255
```

### high-pass filtered images

```python
plt.imshow(img1 - a1, cmap='gray')
```

### low-pass filtered images

```python
plt.imshow(a2, cmap='gray')
plt.show()
```

### hybrid images

```python
plt.imshow(h, cmap='gray')
plt.show()
```
